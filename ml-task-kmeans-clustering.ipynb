```python
# imports for data handling and visualization
import pandas as pd #data manipulation
import matplotlib.pyplot as plt # for plotting
import seaborn as sns # for prettier visualizations
```

```python
#Loading the dataset
df = pd.read_csv("data/Mall_Customers.csv")
```

```python
# Understanding the dataset
df.info()
df.head()
df.describe()
```

```python
#For data preprocessing
from sklearn.preprocessing import LabelEncoder  #for converting categorical values
```

```python
df.isnull().sum() #checking for missing values
```

```python
#Converting 'Gender' to numeric
if 'Gender' in df.columns:
    le = LabelEncoder()
    df['Gender'] = le.fit_transform(df['Gender']) # Male=1, Female=0
```

```python
#Choosing features that are relevant
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
#Displaying the selected features
X.head()
```

```python
#Visualizing the customers by income vs. spending score
plt.figure(figsize=(8,5))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df)
plt.title("Customer Distribution")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score(1-100)")
plt.grid(True)
plt.show()
```

```python
# Imports for KMeans and metrics
from sklearn.cluster import KMeans
```

```python
#Using Elbow Method to find optimal number of clusters (K)
inertia = [] # inertia = sum of squared distances
K_range = range(1,11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
#Plotting the Elbow Curve
plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
# K = 5 is looking good 
k_optimal=5
```

```python
#Fit KMeans and predicting the cluster labels
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
clusters = kmeans.fit_predict(X)
```

```python
#Adding the cluster label to original dataframe
df['Cluster'] = clusters
```

```python
#Visualizing the clusters
plt.figure(figsize=(8,5))
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=clusters, palette='Set1', s=100)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', marker='X', label='Centroids')
plt.title("Customer Segments")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#importing silhouette score
from sklearn.metrics import silhouette_score
```

```python
#Evaluating the clustering performance
score = silhouette_score(X, clusters)
print(f"Silhouette Score for K={k_optimal}: {score:.3f}")
```

```python
# So from the above Silhouette Score of 0.554, we can say that clusters perfomance is 'better' and not 'yikes'
```
